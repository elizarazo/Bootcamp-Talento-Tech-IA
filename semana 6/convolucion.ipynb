```python
import numpy as np
import matplotlib.pyplot as plt

def convolve_manual(X, Y):
  """
  Función para calcular la convolución manual de dos vectores unidimensionales.

  Args:
    X: Primer vector unidimensional.
    Y: Segundo vector unidimensional.

  Returns:
    Vector unidimensional que representa la convolución de X e Y.
  """

  # Verificar longitudes de los vectores
  if len(X) != len(Y):
    raise ValueError("Los vectores deben tener la misma longitud.")

  # Inicializar vector de resultado
  result = np.zeros(len(X) + len(Y) - 1)

  # Realizar la convolución manual
  for i in range(len(X)):
    for j in range(len(Y)):
      result[i + j] += X[i] * Y[j]

  return result

def plot_convolution(X, Y, result):
  """
  Función para visualizar la convolución de dos vectores unidimensionales.

  Args:
    X: Primer vector unidimensional.
    Y: Segundo vector unidimensional.
    result: Vector unidimensional que representa la convolución de X e Y.
  """

  # Crear figura y subfiguras
  fig, axs = plt.subplots(2, 1, figsize=(10, 6))

  # Graficar vectores X e Y
  axs[0].plot(X, label='Vector X')
  axs[0].plot(Y, label='Vector Y')
  axs[0].set_title('Vectores originales')
  axs[0].set_xlabel('Posición')
  axs[0].set_ylabel('Valor')
  axs[0].legend()

  # Graficar resultado de la convolución
  axs[1].plot(result, label='Resultado de la convolución')
  axs[1].set_title('Resultado de la convolución')
  axs[1].set_xlabel('Posición')
  axs[1].set_ylabel('Valor')
  axs[1].legend()

  # Ajustar márgenes y mostrar la figura
  plt.tight_layout()
  plt.show()
```

```python
# Ejemplo de uso
X = [1, 2, 3, 4, 5]
Y = [6, 7, 8, 9, 10]

result_manual = convolve_manual(X, Y)
print(result_manual)
plot_convolution(X, Y, result_manual)

X = [1, 1, 1, 1, 1]
Y = [1, 1, 1, 1, 1]

result_manual = convolve_manual(X, Y)
print(result_manual)
plot_convolution(X, Y, result_manual)
```
