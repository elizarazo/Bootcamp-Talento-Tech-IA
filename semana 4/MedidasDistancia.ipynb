```python
# programa de calcular Distancias
from pylab import *
import numpy as np
import matplotlib.pyplot as plt

#valores de las coord
x1=float(input("Cual es la coordenada x1?"))
x2=float(input("Cual es la coordenada x2?"))
y1=float(input("Cual es la coordenada y1?"))
y2=float(input("Cual es la coordenada y2?"))

print("x1=", x1)
print("x2=", x2)
print("y1=", y1)
print("y2=", y2)

def dist_euclidiana(x1, x2,y1,y2):
  return np.sqrt(((x2-x1)**2)+((y2-y1)**2)) # Return the calculated distance
  print("distancia Eucludiana=", np.sqrt(((x2-x1)**2)+((y2-y1)**2)))

def dist_manhattan(x1, x2,y1,y2):
  return abs(x2-x1)+abs(y2-y1) # Return the calculated distance
  print("distancia manhattan=", abs(x2-x1)+abs(y2-y1))

def dist_chebyshev(x1, x2,y1,y2):
  return max(abs(x2-x1),abs(y2-y1)) # Return the calculated distance
  print("distancia chebyshev=", max(abs(x2-x1),abs(y2-y1)))

def dist_minkowski(x1, x2,y1,y2,p):
  return (abs(x2-x1)**p+abs(y2-y1)**p)**(1/p) # Return the calculated distance
  print("distancia minkowski=", (abs(x2-x1)**p+abs(y2-y1)**p)**(1/p))

def dist_coseno(x1, x2,y1,y2):
  return np.arccos(np.dot((x2-x1),(y2-y1))/(np.linalg.norm(x2-x1)*np.linalg.norm(y2-y1))) # Return the calculated distance
  print("distancia coseno=",
        np.arccos(np.dot((x2-x1),(y2-y1))/(np.linalg.norm(x2-x1)*np.linalg.norm(y2-y1))))
```

```python
euclideana= dist_euclidiana(x1, x2,y1,y2)
manhattan= dist_manhattan(x1, x2,y1,y2)
chebyshev= dist_chebyshev(x1, x2,y1,y2)
minkowski= dist_minkowski(x1, x2,y1,y2,2)
cosine= dist_coseno(x1, x2,y1,y2)

# Store distances in a list
distances = [euclideana, manhattan, chebyshev, minkowski, cosine]
labels = ['Euclidean', 'Manhattan', 'Chebyshev', 'Minkowski', 'Cosine']
plt.bar(labels, distances)
plt.xlabel('Distance Metric')
plt.ylabel('Distance')
plt.title('Distance Metrics')
plt.show()
```
