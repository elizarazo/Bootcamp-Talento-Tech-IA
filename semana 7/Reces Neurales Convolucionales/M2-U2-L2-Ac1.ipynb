```python
import numpy as np

# Definir las señales de entrada y el filtro
x = np.array([2, 1, 0, 1, 2])
h = np.array([1, -1])

# Realizar la convolución
y = np.convolve(x, h, mode='valid')

# Imprimir el resultado
print("Señal de entrada:")
print(x)
print("Filtro:")
print(h)
print("Señal convolucionada:")
print(y)
```
